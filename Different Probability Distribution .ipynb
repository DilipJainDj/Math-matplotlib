```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

## uniform probability distribution

```python
values= np.random.uniform(-10.0 ,10.0 , 10000)
fig  , ax = plt.subplots()
ax.hist(values , 50 , color  = '#F96C10' , label = 'Uniform' , alpha =1)
ax.legend(loc = "center")
ax.set_title("Uniform Probability distribution")
plt.show()
```

## Normal / Gaussian Distribution

```python
from scipy.stats import norm
fig  , ax = plt.subplots()
x  = np.arange(-3,3 , 0.001)
ax.plot(x , norm.pdf(x),label = 'gaussion',linewidth = 5,color ='#F96C10',)
ax.legend(loc="best")
ax.set_title("Normal(Gaussion) Probability distribution")
plt.show()
```

```python
mu= 5.0
sigma = 2.0
fig  , ax = plt.subplots()
values = np.random.normal(mu , sigma , 10000)
ax.hist(values , label='gaussion',linewidth =2 , color ='#F96C10')
ax.legend(loc='best')
ax.set_title("Noraml probability distribution using numpy")
plt.show()
```

## Exponential PDF

```python
x = np.arange(0,10,0.001)
from scipy.stats import expon,binom,poisson
fig  , ax = plt.subplots()
ax.plot(x , expon.pdf(x),linewidth  =2 ,label = "Power low plot", color='#F96C10',)
ax.set_title("Exponential PDF")
ax.legend(loc = "best")
plt.show()
```

## Binomial Probability Function

```python
n,p =10,0.5
x=np.arange(0,10,0.001)
fig  , ax = plt.subplots()
ax.plot(x,binom.pmf(x,n,p) ,linewidth =5,label ="Binomial pmf", color = '#F96C10')
ax.legend(loc = "best")
ax.set_title("Binomial Probability Mass Function")
plt.show()
```

## Poission Probability Mass Function




```python
mu = 500
x = np.arange(400 , 600 , 0.5)
fig  , ax = plt.subplots()
ax.plot(x,poisson.pmf(x,mu),label ="poisson pmf", color = '#F96C10')
ax.legend(loc = "best")
ax.set_title("poisson Probability Mass Function")
plt.show()
```
