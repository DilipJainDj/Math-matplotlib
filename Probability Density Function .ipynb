# Probability Density Function & Some Example

```python
import numpy as np
from scipy.stats import norm
from scipy.stats import expon
from scipy.stats import binom
from scipy.stats import poisson
import matplotlib.pyplot as plt
```

### Uniform Probability distribution

```python
values =  np.random.uniform(-10.0 , 10.0 , 100000)
plt.hist(values , 50)
plt.show()
```

## Normal Probability distribution using 2D line Graph

```python
x=  np.arange(-3 , 3 , 0.001)
plt.plot(x, norm.pdf(x))
```

## Normal Probability distibution function using numpy  random function 

```python
mu = 5.0  # mean of normal function
sigma = 2.0 # difference between number 
values = np.random.normal(mu  , sigma , 10000)
plt.hist(values , 50)
plt.show()
```

## Binomial probability mass function

```python
n,p= 10,0.5
x=  np.arange(0,10,0.001)
plt.plot(x , binom.pmf(x,n,p))
plt.show()
```

### Poission Probability Mass function

```python
mu=500
x = np.arange(400 , 600 , 0.5)
plt.plot(x , poisson.pmf(x , mu))
plt.show()
```
